```python
def geometry_based_friction_model(E1, nu1, sigma_y1, rho1, H1, R_a_1, R_q_1, R_z_1,
                                  E2, nu2, sigma_y2, rho2, H2, R_a_2, R_q_2, R_z_2,
                                  F_n):
    """
    Enhanced Friction Model.

    Parameters:
    - E1, E2: Young's modulus (Pa) of the two materials.
    - nu1, nu2: Poisson's ratio of the two materials.
    - sigma_y1, sigma_y2: Yield strength (Pa) of the two materials.
    - rho1, rho2: Density (kg/m³) of the two materials.
    - H1, H2: Hardness (Pa) of the two materials.
    - R_a, R_q, R_z: Surface roughness parameters (m).
    - F_n: Normal force (N).

    Returns:
    - µ_total: Total friction coefficient.
    - µ_adh: Adhesion contribution.
    - µ_plow: Plowing contribution.
    - A_real_um2: Real contact area (in µm²).
    """
    # 1. Effective Young's Modulus using Hertzian Contact Theory
    E_star = 1 / ((1 - nu1**2) / E1 + (1 - nu2**2) / E2)

    # 2. Average Hardness
    H_avg = (H1 + H2) / 2

    # 3. Real Contact Area Estimation
    A_real_m2 = F_n / H_avg  # Real contact area in square meters

    # Convert A_real to µm²
    A_real_um2 = A_real_m2

    # 4. Adhesion Contribution
    tau = E_star  # Shear strength (Pa), can be refined with empirical data
    µ_adh = tau * A_real_m2 / F_n  # Adhesion friction coefficient (dimensionless)

    # 5. Plowing Contribution

    # Calculate average roughness parameters
    R_z_avg = (R_z_1 + R_z_2) / 2  # Average maximum roughness height (m)
    R_a_avg = (R_a_1 + R_a_2) / 2 + 1e-9  # Average arithmetic mean roughness (m), added small term to avoid division by zero

    # Define shear strength (tau)
    # Option 1: Use a fixed fraction of the effective Young's Modulus
    tau = 0.1 * E_star  # Shear strength (Pa)

    # Option 2: Use yield strength as shear strength (more physically meaningful)
    # tau = min(sigma_y1, sigma_y2)  # Shear strength based on lower yield strength (Pa)

    # Choose one of the above options based on model requirements
    # Here, we'll proceed with Option 1
    # tau = 0.1 * E_star

    # Calculate plowing factor based on roughness
    plowing_factor = R_z_avg / R_a_avg  # Dimensionless

    # Calculate plowing force (N)
    F_plow = tau * A_real_m2 * plowing_factor  # Force due to plowing (N)

    # Calculate plowing friction coefficient (dimensionless)
    µ_plow = F_plow / F_n  # Plowing friction coefficient

    # 6. Total Friction Coefficient
    µ_total = µ_adh + µ_plow

    return µ_total, µ_adh, µ_plow
```
